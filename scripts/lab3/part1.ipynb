```python
import pandas as pd
import matplotlib.pyplot as plt

trace1 = "trace1.csv"
trace2 = "trace2.csv"

trace1 = pd.read_csv(trace1)
trace2 = pd.read_csv(trace2)
```

1. 在不同的 Cache Size( 1KB ~ 64KB) 的条件下, Miss Rate 随 Block Size变化的趋势,收集数据
并绘制折线图。并说明变化原因。至少有4个不同的size大小对应的折线图。(40分)

```python
# Define the column types
column_types = {
    'CacheSize': int,
    'BlockSize': int,
    'Associativity': int,
    'WritePolicy': str,
    'WriteMissPolicy': str,
    'EvictPolicy': str,
    'TotalReads': int,
    'ReadMisses': int,
    'TotalWrites': int,
    'WriteMisses': int,
    'TotalLatency': int,
}

# Apply the column types to trace1 and trace2
trace1 = trace1.astype(column_types)
trace2 = trace2.astype(column_types)
```

```python
# First, let's analyze multiple cache sizes (1KB, 4KB, 16KB, 64KB)
cache_sizes = [1024, 4096, 16384, 65536]  # in bytes
plt.figure(figsize=(12, 8))
for cache_size in cache_sizes:
    plt.figure(figsize=(8, 6))
    
    # Filter data for the current Cache Size
    trace1_filtered = trace1[
        (trace1['CacheSize'] == cache_size) &
        (trace1['Associativity'] == 8) &
        (trace1['WritePolicy'] == '--WriteBack') &
        (trace1['EvictPolicy'] == '--LRU') &
        (trace1['WriteMissPolicy'] == '--WriteAllocate')
    ]
    trace2_filtered = trace2[
        (trace2['CacheSize'] == cache_size) &
        (trace2['Associativity'] == 8) &
        (trace2['WritePolicy'] == '--WriteBack') &
        (trace2['EvictPolicy'] == '--LRU') &
        (trace2['WriteMissPolicy'] == '--WriteAllocate')
    ]
    
    # Calculate Miss Rate
    trace1_filtered['MissRate'] = (trace1_filtered['ReadMisses'] + trace1_filtered['WriteMisses']) / \
                                  (trace1_filtered['TotalReads'] + trace1_filtered['TotalWrites'])
    trace2_filtered['MissRate'] = (trace2_filtered['ReadMisses'] + trace2_filtered['WriteMisses']) / \
                                  (trace2_filtered['TotalReads'] + trace2_filtered['TotalWrites'])
    
    # Plot for the current cache size
    plt.plot(trace1_filtered['BlockSize'], trace1_filtered['MissRate'], 
             label='Trace 1', marker='o')
    plt.plot(trace2_filtered['BlockSize'], trace2_filtered['MissRate'], 
             label='Trace 2', marker='^')
    
    plt.title(f'Miss Rate vs Block Size for Cache Size {cache_size // 1024}KB')
    plt.xlabel('Block Size (bytes)')
    plt.ylabel('Miss Rate')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

2. 在给定 Cache Size 的条件下,如64KB(L1 Cache)或512KB(L2 Cache), Miss Rate 随
Associativity( 1-32) 变化的趋势,收集数据并绘制折线图。并说明变化原因。至少有2、4、8、
16对应的折线图。(40分)​

```python
# Filter data for CacheSize 64KB (65536 bytes)
cache_size = 65536
trace1_filtered = trace1[
    (trace1['CacheSize'] == cache_size) &
    (trace1['WritePolicy'] == '--WriteBack') &
    (trace1['EvictPolicy'] == '--LRU') &
    (trace1['WriteMissPolicy'] == '--WriteAllocate') &
    (trace1['BlockSize'] == 32)
]
trace2_filtered = trace2[
    (trace2['CacheSize'] == cache_size) &
    (trace2['WritePolicy'] == '--WriteBack') &
    (trace2['EvictPolicy'] == '--LRU') &
    (trace2['WriteMissPolicy'] == '--WriteAllocate') &
    (trace2['BlockSize'] == 32)
]

# Group by Associativity and calculate average Miss Rate for all policies
trace1_filtered['MissRate'] = (trace1_filtered['ReadMisses'] + trace1_filtered['WriteMisses']) / \
                              (trace1_filtered['TotalReads'] + trace1_filtered['TotalWrites'])
trace2_filtered['MissRate'] = (trace2_filtered['ReadMisses'] + trace2_filtered['WriteMisses']) / \
                              (trace2_filtered['TotalReads'] + trace2_filtered['TotalWrites'])

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(trace1_filtered['Associativity'], trace1_filtered['MissRate'], marker='o', label='Trace 1')
plt.plot(trace2_filtered['Associativity'], trace2_filtered['MissRate'], marker='o', label='Trace 2')


plt.title('Miss Rate vs Associativity (Cache Size: 64KB)')
plt.xlabel('Associativity')
plt.ylabel('Miss Rate')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

3. 比较 Write Through 和 Write Back、 Write Allocate 和 No-write allocate 的总访问延时的差异。
(20分)​

我们的配置hit latency是3周期，miss penalty是220周期

```python
# Filter trace1 for the given conditions
filtered_trace1 = trace1[
    (trace1['BlockSize'] == 32) &
    (trace1['Associativity'] == 8) &
    (trace1['EvictPolicy'] == '--LRU')
]
# Separate data for each WritePolicy and WriteMissPolicy combination
write_policies = ['--WriteThrough', '--WriteBack']
write_miss_policies = ['--WriteAllocate', '--NoWriteAllocate']

plt.figure(figsize=(10, 6))

# Plot lines for each combination
for write_policy in write_policies:
    for write_miss_policy in write_miss_policies:
        subset = filtered_trace1[
            (filtered_trace1['WritePolicy'] == write_policy) &
            (filtered_trace1['WriteMissPolicy'] == write_miss_policy)
        ]
        if not subset.empty:
            plt.plot(subset['CacheSize'], subset['TotalLatency'], label=f"{write_policy}, {write_miss_policy}", marker='o')

# Add chart details
plt.title('Latency vs Cache Size (Block Size: 32, Associativity: 8, EvictPolicy: LRU)')
plt.xlabel('Cache Size (bytes)')
plt.ylabel('Latency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
